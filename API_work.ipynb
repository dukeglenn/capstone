# Building Out Case Lists

```python
# Input a precedent ID, return an Opinions Cited API pull for that precedent
def opinions_from_precedents(precedent_ids):
  opinions = []
  for precedent in precedent_ids:
    filtered_url = f'https://www.courtlistener.com/api/rest/v4/opinions-cited/?cited_opinion={precedent}'
    headers = {"Authorization": f'Token {api_key}'}
    params = {'limit': 20}

    response = requests.get(filtered_url, headers=headers, params=params)
    #print(response.json())
    if "detail" not in response.json():
      data = response.json()["results"]
      opinions.extend(data)
    else:
      print("Check if out of pulls")
  #print(opinions)
  return opinions
```

```python
# Input a dictionary of Opinions Cited API pulls, return the opinion ids of the citing opinions
def id_from_opinions_cited(opinions_cited):
  ids = []
  for opinion in opinions_cited:
    opinion = opinion.get("citing_opinion").split("/")
    ids.append(opinion[-2])
  return ids
```

```python
# Input a list of precedents, return all opinions connected via citation
# max_depth added to account for runtime, increase in order to get more cases

def opinion_tree(precedent_ids, max_depth = 5):
  if max_depth == 0:
    return precedent_ids

  citing_opinions = opinions_from_precedents(precedent_ids)

  if len(citing_opinions) > 0:
    opinion_ids = id_from_opinions_cited(citing_opinions)
    if set(opinion_ids) == set(precedent_ids):
      return precedent_ids
    else:
      return opinion_tree(opinion_ids, max_depth = max_depth - 1)
  else:
    return precedent_ids
```

```python
# Tutrani v. County of Suffolk, Barile v. Lazzarini, Cabrera v. Rodriguez, Hanford v. Plaza Packaging Corp, Macauley v. Elrac, Inc., DeAngelis v. Kirschner

precedent_ids = [2214438, 6131371, 5938835, 2006666, 5818272, 6070954] # Manually found
```

```python
opinion_ids = opinion_tree(precedent_ids)
opinion_ids
```

```python
# Remove duplicates, show number of unique opinion IDs

opinion_ids = list(set(opinion_ids))
print(f'{len(opinion_ids)} opinions gathered')
```

```python
# Pull the opinions from the IDs gathered

opinions_df = pd.DataFrame(columns=['resource_uri', 'id', 'absolute_url', 'cluster_id', 'cluster', 'author_id',
                                    'author', 'joined_by', 'date_created', 'date_modified', 'author_str',
                                    'per_curiam', 'joined_by_str', 'type', 'sha1', 'page_count',
                                    'download_url', 'local_path', 'plain_text', 'html', 'html_lawbox',
                                    'html_columbia', 'html_anon_2020', 'xml_harvard', 'html_with_citations',
                                    'extracted_by_ocr', 'ordering_key', 'opinions_cited'])
headers = {"Authorization": f'Token {api_key}'}
params = {'limit': 1000}

for id in opinion_ids:
  response = requests.get(f'https://www.courtlistener.com/api/rest/v4/opinions/{id}', headers=headers, params=params)
  if "detail" not in response.json():
    new_opinion = response.json()
    opinions_df.loc[len(opinions_df.index)] = new_opinion
  else:
    print("Out of pulls")

opinions_df.rename(columns={"id":"opinion_id"}, inplace=True)
opinions_df.set_index("opinion_id", inplace=True)
opinions_df
```
